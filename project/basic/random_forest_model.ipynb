#### Data Collection

```python
import pandas as pd
```

```python
FILEPATH = 'rent_apartments.csv'
data = pd.read_csv(FILEPATH)
```

```python
data.head()
```

#### Data Preprocessing

```python
## Transform Categorical Features into Numerical Features
data_encoded = pd.get_dummies(data, columns=['balcony', 'storage', 'parking', 'garage', 'furnished'], drop_first=True)
```

```python
## cat_col = data.select_dtypes(include=['object']).columns.tolist()
## num_col = data.select_dtypes(include=['int64', 'float64']).columns.tolist()
```

```python
data_encoded
```

```python
data_encoded.garden.unique()
```

```python
import re
```

```python
data_encoded.garden[4]
```

```python
## using findAll to extract just the numerical value
re.findall(r'\d+', data_encoded.garden[4])
```

```python
## Take it out of the list and convert to integer
int(re.findall(r'\d+', data_encoded.garden[4])[0])
```

```python
## Apply to the data set
for i in range(len(data_encoded)):
    if data_encoded.garden[i] == 'Not present':
        data_encoded.at[i, 'garden'] = 0
    else:
        data_encoded.garden[i] = int(re.findall(r'\d+', data_encoded.garden[i])[0])
```

```python
data_encoded.head()
```

#### Model Building

```python
true_false_columns = data_encoded.select_dtypes(include=['bool']).columns
data_encoded[true_false_columns] = data_encoded[true_false_columns].astype(int)
```

```python
X = data_encoded[['area', 'bedrooms', 'garden', 'constraction_year', 'balcony_yes',	'storage_yes',
                'parking_yes', 'garage_yes', 'furnished_yes']]
```

```python
y = data_encoded.rent
```

#### Split Dataset

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=32)
```

#### Model Building

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor()
```

```python
rf.fit(X_train, y_train)
```

```python
rf.score(X_test, y_test)
```

#### Hyperparameter Tuning to Improve the Model

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_space = {
    'n_estimators' : [100, 200, 300],
    'max_depth': [3, 6, 9, 12]
}
```

```python
grid = GridSearchCV(RandomForestRegressor(), param_grid=grid_space, cv=5, scoring='r2')
```

```python
model_grid = grid.fit(X_train, y_train)
```

```python
print(f'Best Hyperparameters: {model_grid.best_params_}, Score: {model_grid.best_score_}')
```
